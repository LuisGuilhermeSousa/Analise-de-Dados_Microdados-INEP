## Criando Matrizes de Dados

Abaixo estou criando uma matriz de dados para os anos de 2018 ao 2014, e dentro de cada matriz à uma lista de dados onde contém
os parametros já mencionados nas celúlas anteriores.(IN_ESPECIAL_EXCLUSIVA só foi incluído nos dados do Inep a partir do ano de 2015).




```python
## Matriz de Dados do ano de 2018
Sala_Atend_Esp_2018 = []
Banh_PND_2018 = []
Dep_PND_2018 = []
Esp_Exclusiva_2018 = []

## Matriz de Dados do ano de 2017
Sala_Atend_Esp_2017 = []
Banh_PND_2017 = []
Dep_PND_2017 = []
Esp_Exclusiva_2017 = []

## Matriz de Dados do ano de 2016
Sala_Atend_Esp_2016 = []
Banh_PND_2016 = []
Dep_PND_2016 = []
Esp_Exclusiva_2016 = []

## Matriz de Dados do ano de 2015
Sala_Atend_Esp_2015 = []
Banh_PND_2015 = []
Dep_PND_2015 = []
Esp_Exclusiva_2015 = []

## Matriz de Dados do ano de 2014
Sala_Atend_Esp_2014 = []
Banh_PND_2014 = []
Dep_PND_2014 = []
```

### Comando para adicionar os valores as listas:

Execute está linha de código apenas uma vez. Caso tenha executado mais de uma vez, use o comando abaixo para corrigir o problema.

```python
## Comando para adicionar dados a Matriz 2018
Sala_Atend_Esp_2018.append(df2018.query('(IN_SALA_ATENDIMENTO_ESPECIAL == 1 and NU_ANO_CENSO == 2018)')['IN_SALA_ATENDIMENTO_ESPECIAL'].count())
Banh_PND_2018.append(df2018.query('(IN_BANHEIRO_PNE == 1 and NU_ANO_CENSO == 2018)')['IN_BANHEIRO_PNE'].count())
Dep_PND_2018.append(df2018.query('(IN_DEPENDENCIAS_PNE == 1 and NU_ANO_CENSO == 2018)')['IN_DEPENDENCIAS_PNE'].count())
Esp_Exclusiva_2018.append(df2018.query('(IN_ESPECIAL_EXCLUSIVA == 1 and NU_ANO_CENSO == 2018)')['IN_ESPECIAL_EXCLUSIVA'].count())
Dados2018 = {'Sala de Atendimento Especial':int(Sala_Atend_Esp_2018[0]), 'Banheiro PND':int(Banh_PND_2018[0]), 'Dependências':int(Dep_PND_2018[0]), 'Sala de Aula Exclusiva para Pessoas com Deficiência':int(Esp_Exclusiva_2018[0])}


## Comando para adicionar dados a Matriz 2017
Sala_Atend_Esp_2017.append(df2017.query('(IN_SALA_ATENDIMENTO_ESPECIAL == 1 and NU_ANO_CENSO == 2017)')['IN_SALA_ATENDIMENTO_ESPECIAL'].count())
Banh_PND_2017.append(df2017.query('(IN_BANHEIRO_PNE == 1 and NU_ANO_CENSO == 2017)')['IN_BANHEIRO_PNE'].count())
Dep_PND_2017.append(df2017.query('(IN_DEPENDENCIAS_PNE == 1 and NU_ANO_CENSO == 2017)')['IN_DEPENDENCIAS_PNE'].count())
Esp_Exclusiva_2017.append(df2017.query('(IN_ESPECIAL_EXCLUSIVA == 1 and NU_ANO_CENSO == 2017)')['IN_ESPECIAL_EXCLUSIVA'].count())
Dados2017 = {'Sala de Atendimento Especial':int(Sala_Atend_Esp_2017[0]), 'Banheiro PND':int(Banh_PND_2017[0]), 'Dependências':int(Dep_PND_2017[0]), 'Sala de Aula Exclusiva para Pessoas com Deficiência':int(Esp_Exclusiva_2017[0])}


## Comando para adicionar dados a Matriz 2016
Sala_Atend_Esp_2016.append(df2016.query('(IN_SALA_ATENDIMENTO_ESPECIAL == 1 and NU_ANO_CENSO == 2016)')['IN_SALA_ATENDIMENTO_ESPECIAL'].count())
Banh_PND_2016.append(df2016.query('(IN_BANHEIRO_PNE == 1 and NU_ANO_CENSO == 2016)')['IN_BANHEIRO_PNE'].count())
Dep_PND_2016.append(df2016.query('(IN_DEPENDENCIAS_PNE == 1 and NU_ANO_CENSO == 2016)')['IN_DEPENDENCIAS_PNE'].count())
Esp_Exclusiva_2016.append(df2016.query('(IN_ESPECIAL_EXCLUSIVA == 1 and NU_ANO_CENSO == 2016)')['IN_ESPECIAL_EXCLUSIVA'].count())
Dados2016 = {'Sala de Atendimento Especial':int(Sala_Atend_Esp_2016[0]), 'Banheiro PND':int(Banh_PND_2016[0]), 'Dependências':int(Dep_PND_2016[0]), 'Sala de Aula Exclusiva para Pessoas com Deficiência':int(Esp_Exclusiva_2016[0])}


## Comando para adicionar dados a Matriz 2015
Sala_Atend_Esp_2015.append(df2015.query('(IN_SALA_ATENDIMENTO_ESPECIAL == 1 and NU_ANO_CENSO == 2015)')['IN_SALA_ATENDIMENTO_ESPECIAL'].count())
Banh_PND_2015.append(df2015.query('(IN_BANHEIRO_PNE == 1 and NU_ANO_CENSO == 2015)')['IN_BANHEIRO_PNE'].count())
Dep_PND_2015.append(df2015.query('(IN_DEPENDENCIAS_PNE == 1 and NU_ANO_CENSO == 2015)')['IN_DEPENDENCIAS_PNE'].count())
Esp_Exclusiva_2015.append(df2015.query('(IN_ESPECIAL_EXCLUSIVA == 1 and NU_ANO_CENSO == 2015)')['IN_ESPECIAL_EXCLUSIVA'].count())
Dados2015 = {'Sala de Atendimento Especial':int(Sala_Atend_Esp_2015[0]), 'Banheiro PND':int(Banh_PND_2015[0]), 'Dependências':int(Dep_PND_2015[0]), 'Sala de Aula Exclusiva para Pessoas com Deficiência':int(Esp_Exclusiva_2015[0])}


## Comando para adicionar dados a Matriz 2014
Sala_Atend_Esp_2014.append(df2014.query('(ID_SALA_ATENDIMENTO_ESPECIAL == 1 and ANO_CENSO == 2014)')['ID_SALA_ATENDIMENTO_ESPECIAL'].count())
Banh_PND_2014.append(df2014.query('(ID_SANITARIO_PNE == 1 and ANO_CENSO == 2014)')['ID_SANITARIO_PNE'].count())
Dep_PND_2014.append(df2014.query('(ID_DEPENDENCIAS_PNE == 1 and ANO_CENSO == 2014)')['ID_DEPENDENCIAS_PNE'].count())
Dados2014 = {'Sala de Atendimento Especial':int(Sala_Atend_Esp_2014[0]), 'Banheiro PND':int(Banh_PND_2014[0]), 'Dependências':int(Dep_PND_2014[0])}


## Demonstrativo da Matriz de Dados
print (f'Os dados de 2018 são: {Dados2018}')
print (f'Os dados de 2017 são: {Dados2017}')
print (f'Os dados de 2016 são: {Dados2016}')
print (f'Os dados de 2015 são: {Dados2015}')
print (f'Os dados de 2014 são: {Dados2014}')
```
